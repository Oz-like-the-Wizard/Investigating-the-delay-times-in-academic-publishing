```python
retr_anlys = retr_complete_one.copy()
```

```python
retr_anlys.loc[retr_anlys["Issn"].map(type) == list,"Issn"] = retr_anlys[retr_anlys["Issn"].map(type) == list]["Issn"].map(tuple)
```

```python
retr_analys_vcounts = retr_anlys.groupby("Issn")["Data_Retr"].value_counts(normalize=True)
```

```python
retr_analys_vcounts.sort_values()
```

```python
retr_analys_vcounts[retr_analys_vcounts == 1]
```

```python
retr_analys_vcounts = retr_analys_vcounts.rename("Ratio").reset_index("Data_Retr")
```

```python
# PERFECT jrnls, retrieved succesfully for every year
retr_analys_vcounts[(retr_analys_vcounts.Ratio == 1) & (retr_analys_vcounts.Data_Retr == True)]
```

```python
# WORST jrnls, CANNOT retrieve for a single year
retr_analys_vcounts[(retr_analys_vcounts.Ratio == 1) & (retr_analys_vcounts.Data_Retr == False)]
```

```python
retr_analys_vcounts[(retr_analys_vcounts.Ratio != 1) & (retr_analys_vcounts.Data_Retr == True)].sort_values("Ratio")
```

```python
# In between jrnls, retrieved for some years but not for all!
in_btw_issns = retr_analys_vcounts[(retr_analys_vcounts.Ratio != 1) & (retr_analys_vcounts.Data_Retr == True)].sort_values("Ratio").index.tolist()
```
